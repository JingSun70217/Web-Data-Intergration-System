# Check overlapping of original datasets  

Steps:  
Normalizing songName in 'lyrics.csv';  
Normalizing song in 'SPARQL_1.csv';  
Counting numbers of songs in each datasets separately;  
Calcuting the total number of songs in these two datasets, assume it's A;  
Merging two datasets according to songName;  
Counting the number of songs in merged datasets, assume it's B;  
A - B = C, The gap of A and B is the number of songs which are merged;  
That is C is the number of overlapping.

```python
import pandas as pd
import numpy as np
import re
path = 'C:\\Users\\D074110\\Desktop\\Study\\课\\Web Data Integration\\WDI_TP\\Data collection\\final datasets\\'
```

```python
# Loading 'lyrics.csv'. What does the original 'song' looks like in 'lyrics.csv'
f1 = open(path + 'lyrics.csv',encoding="utf8")
df1=pd.read_csv(f1,sep=',')
df1['song'] 
#There are - inside of songName
#All the words are lowercase. But there are some capitals in another dataset. So we should normalize
```

```python
#Normalizing songName in 'lyrics.csv'
df1['song']=df1['song'].str.replace('-',' ') # Remove - inside of songName
df1['song']=df1['song'].str.title()# Every word in the songName beginning with capital
df1=df1.rename(columns={'song':'songName'}) #Reaname 'song' with 'songName'
df1=df1.rename(columns={'artist':'artistName'})
print(df1.head())
print(df1.shape)
df1['songName'] #Check songName of 'lyrics.csv'
```

```python
#Remove () by using regular expression
artistName = df1['artistName']
artistName = ','.join(str(v) for v in artistName)  # transfer data into string. Because re.sub() expected string or bytes-like object
pattern=re.compile(r'\((.*?)\)')
artistName = pattern.sub('',artistName)
artistName = artistName.replace('-',' ')
artistName = artistName.strip() 
artistName = artistName.split(',') #transfer data into list
artistName = [line.strip() for line in artistName]  #remove skip at the begging and end of string
df1['artistName'] = artistName #It's list now
df1['artistName']=df1['artistName'].str.title()
df1['artistName']
```

```python
# Loading 'SPARQL_1.csv'. What does the original 'song' looks like in 'SPARQL_1.csv'
f2 = open(path + 'SPARQL_1.csv',encoding="utf8")
df2=pd.read_csv(f2,sep='" , "', error_bad_lines=False,warn_bad_lines=False)
df2['songName'] 
#You can see mant problems 
```

```python
#Normalizing songName in SPARQL_1.csv'
#Remove ""
df2 = df2.rename(columns=lambda x: x.replace("'","").replace('"','')).replace(" ","") 
df2['Artistname']=df2['Artistname'].str.replace('"','')
df2['songName']=df2['songName'].str.replace('"','')
df2['songName']=df2['songName'].str.replace(' ,  ,  ,  ,','')#Remove ,  ,  ,  ,
df2['songName']=df2['songName'].str.title()# Every word in the songName beginning with capital
df2=df2.rename(columns={'Artistname':'artistName'})
print(df2.head())
print(df2.shape)
```

```python
#Remove () by using regular expression
songName = df2['songName']
songName = ','.join(str(v) for v in songName)  # transfer data into string. Because re.sub() expected string or bytes-like object
pattern=re.compile(r'\((.*?)\)')
songName = pattern.sub('',songName)
songName = songName.strip() 
songName = songName.split(',') #transfer data into list
songName = [line.strip() for line in songName]  #remove skip at the begging and end of string
songName #It's list now
```

```python
#append data from list to dataframe
df2['songName'] = pd.Series(songName)
df2['songName']=df2['songName'].str.replace('-',' ') 
print(df2.head())
print(df2.shape)
df2['songName']
```

```python
#Remove duplicates value accoeding to 'songs' for each dataset 
#df1 = df1.drop_duplicates('songName',keep='first') 
#df2 = df2.drop_duplicates('songName',keep='first')
print(df1.shape)
print(df2.shape)

df1_len = len(df1)
df2_len = len(df2)
```

```python
#merge two datasets according to 'songName' into a new dataset
#df = pd.merge(df1, df2, how='left', on=['song','songName'])
df = pd.merge(df1, df2, on = 'songName',how = 'outer')
print(df.head())
print(df.shape)
```

```python
#Print the overlapping number
df_len_actual = len(df)
print("After merging,the number of songs is {} ".format(df_len_actual))
df_len_total = df1_len+df2_len
print("The total number of two datasets is {} + {} = {}".format(df1_len,df2_len, df_len_total))
print("That means there are {} songs are verlapping.".format(df_len_total-df_len_actual))
```

```python
#Print the overlapping number
df4_len_actual = len(df4)
print("After merging,the number of songs&artist is {} ".format(df4_len_actual))
df4_len_total = df1_len+df3_len
print("The total number of two datasets is {} + {} = {}".format(df1_len,df3_len, df4_len_total))
print("That means there are {} songs&artist  are verlapping.".format(df4_len_total-df4_len_actual))
```
